```python
# DataFrame 1: data_df
data_df = pd.DataFrame({
    "ID": [1, 2, 3, 4,5],
    "Company": ["TechCorp", "BizSolutions", "WebInnovators", "CreativeWorks","test"]
})

# DataFrame 2: company_data_df
company_data_df = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Company": ["AlphaInc", "BetaTech", "GammaCorp", "DeltaLLC"]
})

# Merge the two DataFrames with suffixes
pd.merge(data_df, company_data_df, on="ID", suffixes=('_left', '_right'))

```

```python
#inner join operation between two DataFrames (data_df and company_data_df) on the "ID" column.
pd.merge(data_df, company_data_df, on="ID", how='inner')
#Only the rows that have matching values in the ID column in both DataFrames will be included in the result.
```

```python
#outer join operation between two DataFrames (data_df and company_data_df) on the "ID" column.
pd.merge(data_df, company_data_df, on="ID", how='outer')
#Non-matching rows from both DataFrames are still included, but with NaN for the missing values from the other DataFrame.
```

Outer Join (how='outer'): Includes all rows from both DataFrames, filling with NaN where there are no matches.
Inner Join (how='inner'): Only includes rows that have matching values in the ID column from both DataFrames.

```python
pd.merge(data_df, company_data_df, on="ID", how='outer',indicator=True)
```

```python
pd.merge(data_df, company_data_df, on="ID", how='left')
```

```python
pd.merge(data_df, company_data_df, on="ID", how='right')
```

## handling missing values
The fillna() function is used to replace missing values (NaN) with a specified value or method. You can replace missing values with a constant (such as 0, the mean, median, or mode), or use various methods like forward fill (ffill) or backward fill (bfill).

```python
df['rating'] = df['rating'].fillna(0)  # Fill missing ratings with 0
df['budget'] = df['budget'].fillna(df['budget'].median())  # Fill missing budget with median value
df['director'] = df['director'].fillna(method='ffill') #Fill with forward fill (previous value)
df['country'] = df['country'].fillna(method='bfill') # Fill with backward fill (next value)
```

```python
# You can use a dictionary to fill different columns with different values:
df = df.fillna({
    # 'rating': df['rating'].mean(),  # Fill missing ratings with mean
    'score': 0,  # Fill missing scores with 0
    'budget': df['budget'].median(),  # Fill missing budget with median
    'votes': df['votes'].mode()[0],  # Fill missing votes with mode (most frequent value)
})
```

```python
df = df.fillna({
    'released': 0,  # Fill missing scores with 0
    'runtime': df['runtime'].median(),  # Fill missing budget with median
    'gross': df['gross'].mode()[0],  # Fill missing votes with mode (most frequent value)
})
```

```python
df['writer'] = df['writer'].fillna(method='ffill') #Fill with forward fill (previous value)
df['star'] = df['star'].fillna(method='bfill') # Fill with backward fill (next value)
df['company'] = df['company'].fillna(method='bfill') # Fill with backward fill (next value)
```

```python
df.isna().sum()
```

```python
# Check for duplicate rows
duplicates = df.duplicated()

# View the rows that are duplicates (True)
print(df[duplicates])

# View the rows that are not duplicates (False)
print(df[~duplicates])
```

```python
# Check for duplicates based on specific columns (e.g., 'name' and 'year')
duplicates = df.duplicated(subset=['name', 'year'])
print(df[duplicates])
```

```python
# # Remove duplicate rows
# df_cleaned = df.drop_duplicates()

# # Remove duplicate rows based on specific columns (e.g., 'name' and 'year')
# df_cleaned = df.drop_duplicates(subset=['name', 'year'])

# Remove duplicates in place
df.drop_duplicates(inplace=True)


# Reset the index after removing duplicates
df.reset_index(drop=True)

# View the cleaned DataFrame
df.duplicated()
```
