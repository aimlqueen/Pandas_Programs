```python
duplicate_row_df = pd.DataFrame({
    'Position': ['sales executive'],
    'Education': ['bachelors'],
    'Experience': [0],
    'Salary': [100],
    'Location': ['UK']
})

# Concatenate the original DataFrame with the duplicate row DataFrame
a = pd.concat([a, duplicate_row_df], ignore_index=True)
```

```python
import numpy as np
import pandas as pd
data=pd.read_csv('dataset/train-data.csv')
data['Mileage']=data['Mileage'].str.replace('kmpl',"milege").fillna('test')
data['Power']=data['Power'].str.replace('bhp',"power").fillna('testpower')
data['New_Price']=data['New_Price'].str.replace('lakh',"new_price").fillna('testprice')
data['Engine']=data['Engine'].str.replace('CC',"engine").fillna('enginetest')
data
```

```python
data['Year']=data['Year'].astype(float)
data.dtypes
```
