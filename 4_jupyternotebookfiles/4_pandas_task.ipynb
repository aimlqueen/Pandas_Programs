```python
import pandas as pd
data=pd.read_csv('dataset/weather_data.csv')
data.head()
```

```python
# group by city
grouped_city=data.groupby('city')
grouped_city#address where it stored
```

```python
for i,j in grouped_city:
    print(i,j)
```

```python
grouped_city.apply(display,include_groups=False)
```

```python
grouped_city.get_group('paris')
```

```python
grouped_city.min()
```

```python
grouped_city.get_group('paris').max()
```

```python
gd=data.groupby(['city','event'])
gd.apply(display)
```

```python
gd.get_group(('newyork','Sunny'))
```

```python
data.pivot(index='day',columns='event')
```

```python
data.pivot_table(index='day',columns='city',values='windspeed')
```
